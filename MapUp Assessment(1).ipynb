### Question 1: Car Matrix Generation

```python
def generate_car_matrix(df):
    # Pivot the DataFrame using id_1 and id_2
    pivoted_df = df.pivot(index='id_1', columns='id_2', values='car')
    
    # Fill NaN values with 0
    pivoted_df = pivoted_df.fillna(0)
    
    # Replace diagonal values with 0
    for i in range(min(pivoted_df.shape)):
        pivoted_df.iloc[i, i] = 0
    
    return pivoted_df
```

```python
# Assuming dataset-1.csv is loaded into a DataFrame called 'df'
# Replace 'dataset-1.csv' with the actual file path if needed

result = generate_car_matrix(df1)
print(result)
```

- This code defines a function called generate_car_matrix that takes a DataFrame df as input. 
- It uses the pivot function to rearrange the DataFrame based on 'id_1' and 'id_2' columns, fills NaN values with 0, and then replaces the diagonal values with 0.

### Question 2: Car Type Count Calculation

```python
def get_type_count(df):
    # Add a new column 'car_type' based on conditions
    df['car_type'] = pd.cut(df['car'], bins=[-float('inf'), 15, 25, float('inf')], labels=['low', 'medium', 'high'])
    
    # Calculate count of occurrences for each car_type category
    type_count = df['car_type'].value_counts().to_dict()
    
    # Sort the dictionary alphabetically based on keys
    type_count = dict(sorted(type_count.items()))
    
    return type_count
```

```python
# Assuming dataset-1.csv is loaded into a DataFrame called 'data'
# Replace 'dataset-1.csv' with the actual file path if needed

result = get_type_count(df1)
print(result)
```

- This Python function get_type_count adds a new column 'car_type' to the DataFrame based on the specified conditions using pd.cut(). 
- It then calculates the count of occurrences for each 'car_type' category using value_counts() and converts it to a dictionary. 
- Finally, the dictionary is sorted alphabetically based on keys before being returned.

### Question 3: Bus Count Index Retrieval

```python
def get_bus_indexes(df):
    # Calculate the mean of the 'bus' column
    mean_bus = df['bus'].mean()
    
    # Identify indices where 'bus' values are greater than twice the mean
    bus_indexes = df[df['bus'] > 2 * mean_bus].index.tolist()
    
    # Sort the indices in ascending order
    bus_indexes.sort()
    
    return bus_indexes
```

```python
# Assuming dataset-1.csv is loaded into a DataFrame called 'df'
# Replace 'dataset-1.csv' with the actual file path if needed

result = get_bus_indexes(df1)
print(result)
```

- This Python function get_bus_indexes first calculates the mean of the 'bus' column using mean(). 
- Then it filters the DataFrame to identify indices where the 'bus' values are greater than twice the mean using boolean indexing. 
- The indices are retrieved using .index.tolist(), sorted in ascending order, and returned as a list.

### Question 4: Route Filtering

```python
def filter_routes(df):
    # Group by 'route' and calculate the average of 'truck' column for each route
    route_avg_truck = df.groupby('route')['truck'].mean()
    
    # Filter routes where the average of 'truck' column is greater than 7
    filtered_routes = route_avg_truck[route_avg_truck > 7].index.tolist()
    
    # Sort the routes in ascending order
    filtered_routes.sort()
    
    return filtered_routes
```

```python
# Assuming dataset-1.csv is loaded into a DataFrame called 'data'
# Replace 'dataset-1.csv' with the actual file path if needed

result = filter_routes(df1)
print(result)
```

- This Python function filter_routes first groups the DataFrame by the 'route' column and calculates the average of the 'truck' column for each route using groupby() and mean(). 
- It then filters routes where the average of the 'truck' column is greater than 7, retrieves the indices (which are route names), converts them to a list, sorts the list in ascending order, and returns the sorted list of routes.

### Question 5: Matrix Value Modification

```python
def multiply_matrix(input_df):
    # Deep copy to avoid modifying the original DataFrame
    modified_df = input_df.copy(deep=True)

    # Apply the specified logic to modify values
    modified_df[modified_df > 20] *= 0.75
    modified_df[modified_df <= 20] *= 1.25

    # Round values to 1 decimal place
    modified_df = modified_df.round(1)

    return modified_df
```

- Then, it applies the specified logic to modify the values in the DataFrame.
- Values greater than 20 are multiplied by 0.75, and values 20 or less are multiplied by 1.25. 
- Finally, the modified DataFrame is rounded to 1 decimal place and returned.

### Question 1: Distance Matrix Calculation

```python
def calculate_distance_matrix(df):
    # Assuming the dataset has columns: 'start_id', 'end_id', 'distance'
    
    # Create a pivot table to represent distances between IDs
    pivot = df.pivot(index='id_start', columns='id_end', values='distance').fillna(0)
    
    # Make the matrix symmetric by adding its transpose
    distance_matrix = pivot + pivot.T
    
    # Replace diagonal values with 0
    distance_matrix.values[[range(distance_matrix.shape[0])]*2] = 0
    
    return distance_matrix
```

```python
# Load the dataset-3.csv into a DataFrame
# Replace 'dataset-3.csv' with the actual file path if needed

# Call the function to calculate the distance matrix
result = calculate_distance_matrix(df3)
print(result)
```

- This function calculate_distance_matrix assumes the dataset has columns 'start_id', 'end_id', and 'distance'. 
- It uses pivot() to create a matrix representing distances between different IDs. 
- Then, it makes the matrix symmetric by adding its transpose and replaces diagonal values with zeros.

### Question 2: Unroll Distance Matrix

```python
def unroll_distance_matrix(distance_matrix):
    # Reset index to convert the distance_matrix into long-form
    distance_matrix = distance_matrix.reset_index()
    
    # Melt the DataFrame to reshape it into a long format
    unrolled_distance = pd.melt(distance_matrix, id_vars='id_start', var_name='id_end', value_name='distance')
    
    # Filter out rows where start_id equals end_id
    unrolled_distance = unrolled_distance[unrolled_distance['id_start'] != unrolled_distance['id_end']]
    
    return unrolled_distance[['id_start', 'id_end', 'distance']]
```

```python
# Assuming 'result' contains the distance matrix DataFrame from Question 1
# Replace 'result' with the actual DataFrame variable
unrolled_data = unroll_distance_matrix(df3)
print(unrolled_data)
```

```python
unrolled_data_df = pd.DataFrame(unrolled_data)
print(unrolled_data_df)
```

- This function unroll_distance_matrix takes the DataFrame generated in Question 1 (assuming it's stored in the variable 'result') and transforms it into a long-format DataFrame containing three columns: 'start_id', 'end_id', and 'distance'. 
- It filters out rows where 'start_id' equals 'end_id'.

### Question 4: Calculate Toll Rate

```python
def calculate_toll_rate(df):
    # Define rate coefficients for each vehicle type
    rate_coefficients = {
        'moto': 0.8,
        'car': 1.2,
        'rv': 1.5,
        'bus': 2.2,
        'truck': 3.6
    }
    
    # Calculate toll rates for each vehicle type and add columns to the DataFrame
    for vehicle, rate in rate_coefficients.items():
        df[vehicle] = df['distance'] * rate
    
    return df
```

```python
import pandas as pd

def calculate_time_based_toll_rates0(df):
    # Convert 'startTime' and 'endTime' columns to datetime objects
    df['startTime'] = pd.to_datetime(df['startTime'])
    df['endTime'] = pd.to_datetime(df['endTime'])
    
    # Extract day names and create time intervals
    df['startDay'] = df['startTime'].dt.day_name()
    df['endDay'] = df['endTime'].dt.day_name()
    df['startHour'] = df['startTime'].dt.hour
    df['endHour'] = df['endTime'].dt.hour
    
    # Define discount factors based on time intervals
    def get_discount_factor(row):
        if row['startDay'] in ['Saturday', 'Sunday']:
            return 0.7
        else:
            if 0 <= row['startHour'] < 10 or 18 <= row['startHour'] <= 23:
                return 0.8
            elif 10 <= row['startHour'] < 18:
                return 1.2
    
    # Initialize vehicle columns with default values
    vehicle_columns = ['moto', 'car', 'rv', 'bus', 'truck']
    for vehicle in vehicle_columns:
        df[vehicle] = 0.0
    
    # Apply discount factors based on time ranges to vehicle columns
    for vehicle in vehicle_columns:
        df[vehicle] += df['distance'] * df.apply(lambda row: get_discount_factor(row), axis=1)
    
    return df[['startDay', 'startTime', 'endDay', 'endTime'] + vehicle_columns]
```
